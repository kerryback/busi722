```python
import plotly.express as px
import pandas as pd
import numpy as np

x = np.arange(0, 24.25, 0.25)

df = None 
for i in range(49):
    y = 40 + 10 * (np.log(1 + x) - np.log(12))
    month = (i/4)*np.ones(len(x))
    size = np.ones(len(x))
    size[48-i] = 10
    d = pd.DataFrame(dict(x=x, y=y, month=month, size=size))
    df = pd.concat((df, d))
fig = px.scatter(df, x="x", y="y", animation_frame="month", size="size")
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 100
fig.update_layout(
    font_size=14,
    yaxis=dict(
        range=(0, 100),
        title="Futures Price",
        title_font_size=16,
    ),
    xaxis=dict(
        title="Months to Maturity",
        title_font_size=16,
        tickvals=np.arange(0, 26, 2)
    ),
    template="plotly_dark"
)
fig.show()
```
