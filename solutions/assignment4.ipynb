2. *Generate plots (using python) of the 1x2 and 2x1 (ratio writing) spreads that were discussed in the Maxum Petroleum case.  Assume that you are short crude and include the short position in each of the plots as we did in class. Suppose the price of crude is $80 and just choose option strikes to illustrate the idea.  We want to choose the strikes so that the option premia are in some relation to each other.   What is the desired relationship?*

The 2x1 spread is to buy an out-of-the-money call and sell two deep out-of-the-money calls.  The 1x2 spread is to sell an out-of-the-money call and buy two deep out-of-the-money calls.  In both cases, we would choose the strikes so that the premium of the deep out-of-the-money call is half that of the other call.

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid")

grid = np.arange(0, 160)
strike1 = 100
strike2 = 120
call1 = np.maximum(grid - strike1, 0)
call2 = np.maximum(grid - strike2, 0)
```

```python
portfolio = - grid + call1 - 2*call2 
plt.plot(grid, portfolio)
plt.xlabel("Crude Price")
plt.ylabel("Hedged Short Position")
plt.title("2 x 1 spread")
plt.show()
```

```python
portfolio = - grid - call1 + 2*call2 
plt.plot(grid, portfolio)
plt.xlabel("Crude Price")
plt.ylabel("Hedged Short Position")
plt.title("1 x 2 spread")
plt.show()
```
