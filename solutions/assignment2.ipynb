*Suppose you sell a 135 call, buy a 145 call, buy a 135 put and sell a 145 put.  You sell at the bid price and buy at the ask.*  

(a)  *Plot the value of the portfolio at the options' maturity, ignoring the premiums paid and received.*

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

stock = np.arange(201)
call135 = np.maximum(stock-135, 0)
call145 = np.maximum(stock-145, 0)
put135 = np.maximum(135-stock, 0)
put145 = np.maximum(145-stock, 0)
portfolio = - call135 + call145 + put135 - put145

plt.plot(stock, portfolio)
plt.xlabel("CVX Price")
plt.ylabel("Option Portfolio Value")
plt.show()
```

(b) *Now adjust the plot for the net premium received or paid.*

```python
call135_bid = 8.25
call145_ask = 2.39
put135_ask = 1.89
put145_bid = 6.45

net = portfolio + call135_bid - call145_ask - put135_ask + put145_bid 

plt.plot(stock, net)
plt.xlabel("CVX Price")
plt.ylabel("Option Portfolio Value Net of Premiums")
plt.show()
```

2. *Consider the two-state example discussed in class.  A stock starts at 100 and will either go up 10% or down 10%.  The risk-free rate is 5%.  Price a call option with a strike of 100.*

```python
ru = 0.1
rd = -0.1
rf = 0.05
S = 100
K = 100

stock = np.array([S*(1+rd), S*(1+ru)])
call = np.maximum(stock-K, 0)
p = (rf-rd) / (ru-rd)
expected_call = (1-p)*call[0] + p*call[1]
call_premium = expected_call / (1+rf)

print(f"Value of call option is ${call_premium:.2f}")
```
